# **7.5.Evaluation metrics:**

* we know that the data we are working  with contains outliers,we didnot drop them because if we do so, we may loose out important trends/patterns in the data.
*Decision Trees or any tree based algorithm that we will use here are known to handle outliers.Hence we can use RMSE as the evaluations metrics.
*Since RHSE penalises outliers a lot,this is good metric to check whether or not the model has learnt all the trends/patterns in the data.
*In addition to RMSE,we can use R2 score to make the results more explainable to a larger audience.

```python
#defining rmse evalution metric
def rmse(actual,predicted):
  mse=mean_squared_error(actual,predicted)
  rmse=np.sqrt(mse)
  return rmse
```
